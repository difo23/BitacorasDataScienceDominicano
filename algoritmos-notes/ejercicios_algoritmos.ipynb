# Problema 11: Correos

Se desea encontrar una solución informática para ordenar los correos que se acumulan en una oficina durante un día de servicio, son aproximadamente un millón de correos por día. Los correos para nuestro caso son representados por un par de valores idCiudad e idDireccion. Los valores que puede tomar idCiudad son de 1 a 20 y los valores que puede tomar idDireccion son de 0 a 10000. 

Ejemplo de un correo seria: 
* idCiudad: idDireccion 
* ‘2’:9000 
* ‘1’: 9000
* ‘19’: 0

• #1 Se debe implementar un generador de correos aleatorio para simular el millón de correos acumulados en la oficina.(Los correos se pueden repetir hacia la misma ciudad y la misma dirección). 

• #2 Se debe organizar los correos por casilleros, los cuales tendrán como identificador en idCiudad. Para esta tarea se debe implementar el algoritmo BucketSort. 

• #3 Los correos dentro de los casilleros se deben ordenar por el idDireccion de menor a mayor. Se debe utilizar para esta tarea el algoritmo HeapSort. 

• Los correos individuales deben ser representados por llave-valor (Investigar diccionarios en Python).

```
# Librerisas y constantes
from random import randrange

NUMBER_OF_CITIES = 20
NUMBER_OF_ADDRESS = 10000
```

```
# Problema 1: Crear correos ramdon

def gen_correos(n):
    '''Genera correos aleatorios a partir de un n dado.

    Input: N numero de correos a crear.
    Output: correos generados aleatoriamente diccioanrio Ejem: {1:20}
    '''

    emails = build_buckets()

    for email in range(n):
        # creo un correo de forma aleatoria
        city = randrange(1, NUMBER_OF_CITIES + 1)
        address = randrange(0, NUMBER_OF_ADDRESS + 1)
        # Agrego al casillero correspondiente el correo creado
        emails[city].append(address)
        # Ordeno las direcciones dentro de cada casillero.
    emails= ordenar(emails)
    return emails
```

```
def ordenar(buckets):
  for  key  in buckets:
     buckets[key] = heapSort(buckets[key])
  return buckets
```

```
# Problema 2: Ordenar con el algoritmo BucketSort, cada casillero es una ciudad.
def build_buckets(n=NUMBER_OF_CITIES):
  '''Crea un numero n de casilleros para ordenar

    Input: N numero de casilleros a crear.
    Output: Envia un diccionario, para representar casilleros
  '''
  buckets = {}
  for i in range(1, n+1):
      buckets[i] = []
  return buckets
```

```
# Problema 3:  Ordenar con el algoritmo HeapSort dentro 
# de cada casillero, menor a mayor ciudad.
def heapSort(lst):

    for start in range(int((len(lst)-2)/2), -1, -1):
      siftdown(lst, start, len(lst)-1)
 
    for end in range(len(lst)-1, 0, -1):
      lst[end], lst[0] = lst[0], lst[end]
      siftdown(lst, 0, end - 1)
    return lst
 
def siftdown(lst, start, end):
  root = start
  while True:
        child = root * 2 + 1
        if child > end: break
        if child + 1 <= end and lst[child] < lst[child + 1]:
            child += 1
        if lst[root] < lst[child]:
            lst[root], lst[child] = lst[child], lst[root]
            root = child
        else:
             break
```
