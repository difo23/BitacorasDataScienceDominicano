### 4. Revertir el orden de las filas en un DF

```python
drinks = pd.read_csv('./drinks.csv')
```

```python
drinks.head()
```

```python
drinks.loc[::-1].head()
```

```python
drinks.loc[::-1].reset_index(drop=True).head()
```
